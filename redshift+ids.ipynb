```python
# this notebook is for data mining and comparing files

# import modules + halo ids, define functions + constants
import tangos
import numpy as np
import matplotlib.pyplot as plt
import pynbody
import timeit
import pickle
```

```python
# get groups where there are >0 stars and <100 gas particles in each sim from amiga.stat Files 
sim_names = ['cptmarvel.cosmo25cmb.4096g5HbwK1BH', 'elektra.cosmo25cmb.4096g5HbwK1BH', 'rogue.cosmo25cmb.4096g5HbwK1BH', 'storm.cosmo25cmb.4096g5HbwK1BH']

for s in range(len(sim_names)):
    file = f"Stat_Files/{sim_names[s]}/{sim_names[s]}.004096/{sim_names[s]}.004096.M200.amiga.stat"
    f = np.genfromtxt(file, skip_header = 2, dtype=float, encoding='utf-8', usecols = (0,2,3,5))

    # define columns from stat file
    grp_ids = f[:,0]
    N_gas = f[:,1]
    N_star = f[:,2]
    Mvir = f[:,3]
    n = len(Mvir)
    
    grp = []
    Mvir_arr = []
    Ngas_arr = []

    # find groups in which there are stars + enough gas particles present
    for i in range(n):
        if N_star[i] > 0:
            continue
    
        if N_gas[i] < 100:
            continue
    
        grp.append(grp_ids[i])
        Mvir_arr.append(Mvir[i])
        Ngas_arr.append(N_gas[i])
    
    print(f"sim: {sim_names[s]}: grps {grp} Ngas {Ngas_arr}")

```
